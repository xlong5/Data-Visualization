```python
import pandas as pd
import matplotlib.pyplot as pls
```

```python
Quantities = pd.read_csv('/Users/xiaoyanglong/Downloads/store_data.csv', index_col='State')
print(Quantities)
print(Quantities.columns)
```

```python
Illnois = Quantities[Quantities.index == 'Illinois']
Illnois
```

```python
Illnois.plot.area(y=['Sales'])
plt.title('Sales Amount in Illnois', fontsize = 16, fontweight = 'bold')
plt.xlabel('Sales Record', fontsize=16, fontweight='bold')
plt.ylabel('Sales Amount', fontsize=16, fontweight='bold')
plt.show()

```
