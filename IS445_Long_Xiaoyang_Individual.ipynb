```python
# What is the "name" of the dataset? Best Selling Nintendo Switch Video Games
# Where did you obtain it? I obtain it from Kaggle
# Where can we obtain it? (i.e., URL) https://www.kaggle.com/datasets/kabhishm/best-selling-nintendo-switch-video-games
# What is the license of the dataset? CC0: Public Domain
#What are we allowed to do with it? Yes we're allowed
#How big is it in file size and in items? 3kb
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('downloads/best_selling_switch_games.csv')
dataset
```

```python
# Best Selling Games by Developer
# dataset['count'] = dataset.sort_values(by=['developer', 'copies_sold'],ascending = [True, False]).groupby('developer').cumcount() + 1
# dataset
dataset['count'] = dataset.groupby('developer')['developer'].transform('count')
dataset
```

```python
developer_df = dataset.sort_values(by=['count'], ascending=False)
developer_df = developer_df.drop_duplicates(['developer']).reset_index()
developer_df
```

```python
plt.plot(figsize = (50, 50))
plt.pie(developer_df['count'], labels=developer_df['developer'], autopct="%.1ff%%")
plt.legend(developer_df['developer'], loc="upper left")
```

```python
from collections import Counter

publisher_raw = dataset['publisher'].dropna().tolist()
publisher_list = []

# Create a list of all publisher in the dataframe
for publisher in publisher_raw:
    publisher = publisher.split(", ")
    for p in publisher:
        publisher_list.append(p)


publisher_counts = Counter(publisher_list)
publisher_dataset = pd.DataFrame.from_dict({'publisher': list(publisher_counts.keys()),'Count': list(publisher_counts.values())})

print(publisher_dataset.head())
```

```python
# Create a DataFrame for the given table
data = {'publisher': ['Nintendo', 'The Pokémon CompanyNintendo', 'Capcom', '.mw-parser-output .plainlist ol,.mw-parser-output .plainlist ul', 'JP: Konami'],
        'Count': [47, 5, 2, 1, 1]}
df = pd.DataFrame(data)

# Create a pie chart using matplotlib
plt.figure(figsize=(8,8))
plt.pie(df['Count'], labels=df['publisher'], autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Distribution of Publishers')
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.plot(year_df['year'],year_df['copies_sold']/1000000, color = 'grey', linestyle='dotted')
plt.ylabel('Million', color = 'grey')
plt.xlabel('Year', color = 'grey')
plt.yticks(fontsize = 12, color = 'grey')
plt.xticks(fontsize = 12, color = 'grey')
plt.title('Number of Games Sold in 2018 - 2022',fontsize = 15, color = 'grey')
plt.show()
```
