```python
# What is the "name" of the dataset? Best Selling Nintendo Switch Video Games
# Where did you obtain it? I obtain it from Kaggle
# Where can we obtain it? (i.e., URL) https://www.kaggle.com/datasets/kabhishm/best-selling-nintendo-switch-video-games
# What is the license of the dataset? CC0: Public Domain
#What are we allowed to do with it? Yes we're allowed
#How big is it in file size and in items? 3kb
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('downloads/best_selling_switch_games.csv')
dataset.head(10)
```

#### Top 10 Best Selling Games

```python
# Number of games sold from 2018 to 2022
dataset['as_of'] = pd.to_datetime(dataset['as_of'])
dataset.head(10)
```

```python
### Separate out year
dataset['year'] = dataset['as_of'].dt.year.astype('str')
dataset.head(10)
```

```python
year_df = dataset.groupby('year').sum().reset_index()
year_df.head(10)
```

```python
plt.figure(figsize=(10,10))
plt.plot(year_df['year'],year_df['copies_sold']/1000000, color = 'grey', linestyle='dotted')
plt.ylabel('Million', color = 'grey')
plt.xlabel('Year', color = 'grey')
plt.yticks(fontsize = 12, color = 'grey')
plt.xticks(fontsize = 12, color = 'grey')
plt.title('Number of Games Sold in 2018 - 2022',fontsize = 15, color = 'grey')
plt.show()
```

```python
# Separate publish time by month, discover the most popular publishing month
dataset['month'] = dataset['as_of'].dt.month.astype('str')
dataset.head(10)
```

```python
month_df = dataset.groupby('month').sum().reset_index()
month_df
```

```python
plt.plot(figsize = (20, 20))
plt.pie(month_df['copies_sold'], labels = month_df['month'], autopct='%.0f%%')
plt.title("Popular Publishing Months Report", color = 'grey')
plt.show()
# Then we can find out September is the most popular publishing month, then it is December and March
```

```python
# Top 5 Number of copies sold by publishers
publisher_df = dataset.groupby('publisher').sum().sort_values(by=['copies_sold'], ascending = False).reset_index().head(5)
publisher_df
```

```python
plt.plot(figsize = (15,15))
plt.pie(publisher_df['copies_sold'], labels = publisher_df['publisher'], autopct='%.0f%%', colors=['lightblue', 'yellow', 'blue', 'green', 'purple'])
plt.title("Top 5 Publishers", color = 'grey')
plt.show()
```
